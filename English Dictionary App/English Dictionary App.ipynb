# English Dictionary App

```python
##  By using json package

import json

def addict(tmp_dict):
    with open('words.txt', 'w') as dic:
        json.dump(tmp_dict, dic)
    
    
def checkdict():
    with open('words.txt', 'r') as dic:
        tmp_dict = json.load(dic)
    return tmp_dict

def neword():
    nw = input("Enter new word : ")
    nwm = input("Enter the meaning of word : ")
    dic = checkdict()
    dic[nw] = nwm
    addict(dic)
    print("Data entry is successfully done in words.txt")
    Main_Menu()

    
def find():
    fin = input("Enter word to find meaning : ")
    dic = checkdict()
    if fin in dic:
        print('the meaning of word ',fin, 'is',dic[fin])
    else:
        print("The word doesn't exist!")
        find()
    Main_Menu()


def update():
    upd = input("Enter word to update meaning : ")
    updm = input("Enter meaning to be updated : ")
    dic = checkdict()
    if upd in dic:
        dic[upd] = updm
        addict(dic)
        print('The word successfully updated')
        print('the new meaning of word ',upd, 'is',dic[upd])
    else:
        print("The word doesn't exist! impossible to update!")
    Main_Menu()
    
def Main_Menu():
    X = input('''Select the item the operation to perform on dictionary 
    press 1 for new word meaning adding
    press 2 for finding the meaning of word
    press 3 for updating of meaning
    press 4 for exit : 
    ''')
    if X=='1':
        neword()
    elif X=='2':
        find()
    elif X=='3':
        update()
    elif X=='4':
        exit()
    else:
        print('Please enter valid number')
        Main_Menu()  
        
        
Main_Menu()
```
